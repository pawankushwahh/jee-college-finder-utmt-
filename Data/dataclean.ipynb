```python
import pandas as pd

# Read the CSV
df = pd.read_csv("merged_jee_cutoff_2018_2025.csv")

# Ensure Round is numeric
df["Round"] = pd.to_numeric(df["Round"], errors="coerce")

# Grouping columns (everything except ranks and round)
group_cols = [
    "Institute",
    "Academic Program Name",
    "Quota",
    "Seat Type",
    "Gender",
    "Year"
]

# Keep the row having the maximum Round in each group
cleaned_df = df.loc[df.groupby(group_cols)["Round"].idxmax()]

# Sort for readability
cleaned_df = cleaned_df.sort_values(
    ["Year", "Institute", "Academic Program Name"]
)

# Save
cleaned_df.to_csv("cleaned_data.csv", index=False)

print(cleaned_df.head())
```
